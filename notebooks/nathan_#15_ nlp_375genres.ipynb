```python
from corextopic import corextopic as ct
from sklearn.feature_extraction.text import CountVectorizer
model = ct.Corex(
    n_hidden = 20 #Nombre de topics à détecter
)
vectorizer= CountVectorizer()
X= vectorizer.fit_transform(list_of_tags)
vocab = vectorizer.get_feature_names_out() #Ensemble des mots uniques
model = model.fit(X, words = vocab)
### Affichage
for i, topic_ngrams in enumerate(model.get_topics(n_words=10)):
    topic_ngrams = [ngram[0] for ngram in topic_ngrams if ngram[1] > 0]
    print("\nTopic %d :"%i)
    print(', '.join(topic_ngrams))    
```

# LDA

## Entrainement du modèle

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer().fit(list_of_tags)

data_vectorized = vectorizer.transform(list_of_tags)

lda_model = LatentDirichletAllocation(n_components=15).fit(data_vectorized)

def print_topics(model, vectorizer):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([vectorizer.get_feature_names_out()[i]
                        for i in topic.argsort()[:-10 - 1:-1]])
        

print_topics(lda_model, vectorizer)
```

> On obtient ainsi nos 10 topics, qui sont bien représentatif.
