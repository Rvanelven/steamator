## Test simple modèle

```python
X = df.drop(columns=['steam_appid', 'pub_nb_games', 'owner_estimated','release_date','revenu','name','developer','publisher', 'owner_median', 'rating', 'popularity', 'nb_review', 'recommended'])
```

```python
y = df['owner_estimated']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state=0)
```

```python
X.columns
```

```python

num_transformer = make_pipeline(SimpleImputer(), StandardScaler())
num_col = make_column_selector(dtype_include=['float64','int64'])

cat_transformer = OneHotEncoder()
cat_col = make_column_selector(dtype_include=['object','bool'])

preproc_basic = make_column_transformer(
    (num_transformer, num_col),
    (cat_transformer, cat_col))
```

```python
pipe = make_pipeline(preproc_basic, Ridge())
```

```python

pipe.fit(X_train,y_train)

# Make predictions
pipe.predict(X_test)

# Score model
pipe.score(X_test,y_test)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaler.transform(X_train)
scaler.transform(X_test)
```

```python
# instanciate TPOTClassifier
tpot = TPOTRegressor(generations=1, population_size=50, verbosity=2, scoring='r2', n_jobs=-1, cv=2)
# process autoML with TPOT
tpot.fit(X_train, y_train)

# print score
print(tpot.score(X_test, y_test))
```

```python
tpot.export(os.path.join(os.getcwd(),'tpot_steamator.py'))
```
