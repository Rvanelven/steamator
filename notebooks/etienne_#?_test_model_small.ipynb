# Def X et y

```python
X = data.drop(columns=['steam_appid','name', 'top_5_tags', 'nb_review', 'owner_estimated', 'rating', 'popularity', 'average_playtime', 'median_playtime'])
y = data.owner_estimated
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Scale

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaler.transform(X_train)
scaler.transform(X_test)
```

# Def model

```python
forest = RandomForestRegressor(n_estimators=100, max_leaf_nodes=1000, max_depth=50)
KNN = KNeighborsRegressor(n_neighbors=10)
lasso = linear_model.Lasso(max_iter=5000,positive=True, fit_intercept=False, )
XGB = XGBRegressor(max_depth=10, n_estimators=100, learning_rate=0.1)
GBR = GradientBoostingRegressor(
    n_estimators=100, 
    learning_rate=0.1,
    max_depth=3
)
```

```python
model = CatBoostRegressor()
model.fit(X_train, y_train)
cross_val_score = cross_val_score(model, X_test, y_test, cv=10, scoring='r2').mean()
```

```python
cross_val_score
```

```python
ensemble = StackingRegressor(
    estimators = [('XGB', XGB), ('forest', forest), ("lasso", lasso)],
    final_estimator = GBR,
                n_jobs=-1)
```

```python
pipe_ensemble = make_pipeline(ensemble)
pipe_ensemble.fit(X_train, y_train)
```
