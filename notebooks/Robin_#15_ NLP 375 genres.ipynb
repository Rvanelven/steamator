## Creation d'une colonne TOP5 tag pour chaque ligne, comportant le nom des 5 colonnes avec le + de tag dans l'ordre croissant

```python
df3=pd.DataFrame(df2.columns.values[np.argsort(-df2.values, axis=1)[:, :5]], columns = ['1st Tag', '2nd Tag', '3rd Tag', '4th Tag', '5th Tag'])
```

```python
df3.shape
```

```python
df3
```

```python
df3['top_5_tags']=df3[['1st Tag', '2nd Tag', '3rd Tag', '4th Tag', '5th Tag']].agg(';'.join, axis=1)
```

```python
df3
```

```python
df4 = df5[['steam_appid', 'name']]
```

```python
df4
```

```python
df_final = df4.join(df3, rsuffix='top_')
```

```python
df_final =df_final.drop(columns=['1st Tag', '2nd Tag', '3rd Tag', '4th Tag', '5th Tag'])
```

```python
df_final
```

> On va ensuite enlever les ; et les remplacer par un espace afin de comptabiliser le nombre de fois où chaque tag revient

```python
df_final.top_5_tags = df_final.top_5_tags.str.replace(';',' ')
```

```python
df_final
```

```python
d=dict()
for line in df_final['top_5_tags']:
    line = line.strip()
    line = line.lower()
    words = line.split(" ")

    for word in words:
        if word in d:
            d[word]=d[word] +1
        else:
            d[word]=1
        
for w in sorted(d, key=d.get, reverse=True):
    print(w, d[w])
    
len(d)
```
