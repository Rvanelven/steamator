```python
def create_dataframe(start, end):
    liste_nom = []
    liste_url = []
    liste_id = []
    for i in range(start, end):
        nom = data.name[i]
        lien = data.screenshots[i]
        resultat = []
        cursor = 30
        id_ = data.steam_appid[i]
        while lien[cursor] != "'":
            resultat.append(lien[cursor])
            cursor += 1
        str1 = ''.join(resultat)
        liste_nom.append(nom)
        liste_url.append(str1)
        liste_id.append(id_)
        test1_1 = pd.DataFrame(liste_nom)
        test1_2 = pd.DataFrame(liste_url)
        test1_3 = pd.DataFrame(liste_id)
        dataframe = pd.concat([test1_1, test1_2, test1_3], axis=1)
    return dataframe
```
