```python
import pandas as pd
```

```python
df1 = pd.read_csv('raw_data/steam.csv')
df2 = pd.read_csv('raw_data/steam_description_data.csv')
df3 = pd.read_csv('raw_data/steam_media_data.csv')
df4 = pd.read_csv('raw_data/steam_requirements_data.csv')
df5 = pd.read_csv('raw_data/steam_support_info.csv')
df6 = pd.read_csv('raw_data/steamspy_tag_data.csv')
```

```python
df1 = df1.rename(columns={'appid': 'steam_appid'})
```

```python
df1_1 = df1.merge(df2, on='steam_appid', how='inner')
```

```python
df1_2 = df1_1.merge(df3, on='steam_appid', how='inner')
```

```python
df1_3 = df1_2.merge(df4, on='steam_appid', how='inner')
```

```python
df1_4 = df1_3.merge(df5, on='steam_appid', how='inner')
```

```python
df6 = df6.rename(columns={'appid': 'steam_appid'})
```

```python
df1_5 = df1_4.merge(df6, on='steam_appid', how='inner')
```

```python
df1_5.to_csv(r'data_full.csv', index = False)
```
